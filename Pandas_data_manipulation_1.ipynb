Introduction to Pandas:
Pandas is an open-source library built on top of the Python programming language. It provides easy-to-use data structures and data analysis tools for working with structured data.

The primary data structures in Pandas are Series (1-dimensional) and DataFrame (2-dimensional), which allow you to store and manipulate data in tabular form, similar to a spreadsheet or database table.

Key Features of Pandas:

Data manipulation: Pandas provides a wide range of functions and methods for manipulating data, including filtering, sorting, grouping, merging, and reshaping.
Data cleaning: Pandas offers tools for handling missing data, converting data types, removing duplicates, and performing other data cleaning tasks.
Data analysis: Pandas supports statistical and mathematical operations, time series analysis, and other analytical tasks.

Integration with other libraries: Pandas integrates well with other Python libraries such as NumPy, Matplotlib, and Scikit-learn, allowing seamless data analysis and visualization.

Getting Started with Pandas:
Installation: You can install Pandas using pip, the Python package manager. Simply run pip install pandas in your terminal or command prompt.
Importing Pandas: Once installed, you can import Pandas into your Python script or Jupyter Notebook using import pandas as pd.

Basic Operations with Pandas:
Creating DataFrames: You can create a DataFrame from various data sources, including lists, dictionaries, NumPy arrays, CSV files, Excel files, SQL databases, and more.

Accessing Data: Pandas provides methods for accessing and retrieving data from DataFrames, including indexing, slicing, and boolean indexing.

Manipulating Data: You can modify data in a DataFrame by adding or removing columns, updating values, and performing transformations.

Analyzing Data: Pandas offers functions for calculating summary statistics, aggregating data, and generating visualizations to gain insights into your data.

Example Code:
Here's a simple example demonstrating how to create a DataFrame in Pandas and perform basic operations:

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']}
df = pd.DataFrame(data)

# Displaying the DataFrame
print("DataFrame:")
print(df)

# Accessing data
print("\nAccessing data:")
print(df['Name'])  # Accessing a column
print(df.iloc[0])  # Accessing a row by index

# Adding a new column
df['Gender'] = ['Female', 'Male', 'Male', 'Male']
print("\nDataFrame with new column:")
print(df)

# Calculating summary statistics
print("\nSummary statistics:")
print(df.describe())
```

Data Preparation:

dropna(): Removes missing or NaN values from a DataFrame.

fillna(): Fills missing values in a DataFrame with specified values.

astype(): Converts data types of columns in a DataFrame.

get_dummies(): Converts categorical variables into dummy/indicator variables (one-hot encoding).

Data Exploration:

head(), tail(): Displays the first or last n rows of a DataFrame.

info(): Provides a concise summary of a DataFrame, including data types and non-null counts.

describe(): Generates descriptive statistics (count, mean, std, etc.) for numeric columns.

value_counts(): Counts the occurrences of unique values in a column.
Data Manipulation:

merge(), concat(): Combines DataFrames by merging or concatenating them.

groupby(): Groups data in a DataFrame by one or more columns for aggregation.

apply(): Applies a function along an axis of the DataFrame.

map(), replace(): Maps values of a column to new values using a dictionary or function.
Feature Engineering:

cut(), qcut(): Discretizes continuous variables into categorical bins.

shift(): Shifts index by a desired number of periods, useful for creating lag features.

rolling(): Applies a rolling window calculation to data (e.g., rolling mean, sum, etc.).

diff(): Calculates the difference between consecutive elements in a DataFrame.

Data Visualization:

plot(): Generates various types of plots (line, bar, scatter, histogram, etc.) directly from a DataFrame.
boxplot(), hist(): Plots boxplots and histograms for visualizing distribution and spread of data.
heatmap(): Creates a heatmap to visualize correlation matrices or other 2D data arrays.

pairplot(): Plots pairwise relationships between variables in a DataFrame.

bar(), pie(): Plots bar and pie charts for categorical data.
Data Sampling and Splitting:

sample(): Randomly samples rows or columns from a DataFrame.

train_test_split(): Splits data into training and testing sets for model evaluation.

stratify(): Ensures that class proportions are preserved during data splitting.
Handling Time Series Data:

to_datetime(): Converts a column to datetime format.
resample(): Aggregates time series data over different time periods.

shift(): Shifts time series data forward or backward in time.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, np.nan, 40],
        'City': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
        'Gender': ['Female', 'Male', 'Male', 'Male']}
df = pd.DataFrame(data)

# Data Preparation
print("Data Preparation:")
print("Original DataFrame:")
print(df)
print("\nDrop rows with missing values:")
print(df.dropna())
print("\nConvert 'Age' column to int (handling missing values):")
df['Age'] = df['Age'].fillna(0).astype(int)
print(df)

# Data Exploration
print("\nData Exploration:")
print("First 2 rows:")
print(df.head(2))
print("\nSummary info:")
print(df.info())
print("\nDescriptive statistics:")
print(df.describe())
print("\nValue counts for 'City' column:")
print(df['City'].value_counts())

# Data Manipulation
print("\nData Manipulation:")
print("Merging two DataFrames:")
df2 = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Salary': [50000, 60000]})
print(pd.merge(df, df2, on='Name'))
print("\nGroup by 'City' and calculate average age:")
print(df.groupby('City')['Age'].mean())

# Feature Engineering
print("\nFeature Engineering:")
print("Discretize 'Age' column into bins:")
df['Age_bins'] = pd.cut(df['Age'], bins=3, labels=['Young', 'Middle-aged', 'Old'])
print(df)
print("\nCreate a lag feature for 'Age' column:")
df['Age_lag'] = df['Age'].shift(1)
print(df)

# Data Visualization
print("\nData Visualization:")
print("Histogram of 'Age' column:")
df['Age'].hist()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
print("Boxplot of 'Age' column:")
df.boxplot(column='Age')
plt.ylabel('Age')
plt.title('Boxplot of Age')
plt.show()
print("Bar chart of 'City' counts:")
df['City'].value_counts().plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Count')
plt.title('City Counts')
plt.show()
```

```python
#multi-indexing
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
```

```python
print("\nTuple pairs after the zip and list command\n",'-'*45, sep='')
print(hier_index)
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index)
print("\nIndex hierarchy\n",'-'*25, sep='')
print(hier_index)

```

```python
print("\nIndex hierarchy type\n",'-'*25, sep='')
print(type(hier_index))
```
